## 1. Importing the relevant modules


```python
from matplotlib import pyplot as plt
import numpy as np
import random

# set seed for random module to ensure reproducibility
random.seed(a=253)
```

## 2. Generating example data
Two variables that have been observed over a year in monthly intervals
- Y1
- Y2

```python
x = range(1,13)
month_names = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Okt", "Nov", "Dec"]

y1 = [random.random() for i in range(1,13)]

y2 = [val*1.3 + random.random() for val in y1]
```

## 3. Some examples

#### First example: 
A simple line graph to show the longitudinal development of variables Y1 an Y2

```python
# Initializing a figure object
fig1, ax_fig1 = plt.subplots()

# line for Y1 with red color and dots as markers
plt.plot(x, y1, color="red", marker="o")

# Showing the figure
plt.show()
```

Adding the second line by repeating the *plt.plot()* command

```python
# Initializing a figure object
fig1, ax_fig1 = plt.subplots()

# line for Y1 with red color and dots as markers
plt.plot(x, y1, color="red", marker="o")

# line for Y2 with red color and dots as markers
plt.plot(x, y2 , color ="green", marker ="o")

# Showing the figure
plt.show()
```

Adding title, axis labels and legend

```python
# Initializing a figure object
fig1, ax_fig1 = plt.subplots()

# line for Y1 with red color and dots as markers
plt.plot(x, y1, color="red", marker="o")

# line for Y2 with red color and dots as markers
plt.plot(x, y2 , color ="green", marker ="o")

# Assigning a figure title and axis labels
plt.title("Two lines in one graph")
plt.xlabel("Amazing X-axis")
plt.ylabel("Incredible Y-axis")

# Adding a legend, location can be chosen between 0 and 10 (0 being the default: "automatic best option")
plt.legend(["Y1","Y2"], loc=1)


# Showing the figure
plt.show()
```

Editing the x-Axis

```python
# Initializing a figure object
fig1, ax_fig1 = plt.subplots()

# line for Y1 with red color and dots as markers
plt.plot(x, y1, color="red", marker="o")

# line for Y2 with red color and dots as markers
plt.plot(x, y2 , color ="green", marker ="o")

# Assigning a figure title and axis labels
plt.title("Two lines in one graph")
plt.xlabel("Amazing X-axis")
plt.ylabel("Incredible Y-axis")


# Adding a legend, location can be chosen between 0 and 10 (0 being the default: "automatic best option")
plt.legend(["Y1","Y2"], loc=1)

# Editing the x-axis
ax_fig1.set_xticks(x)
ax_fig1.set_xticklabels(month_names)

# Shwoning the figure
plt.show()

fig1.savefig("Awesome_figure.png")

#plt.close("all")
```

#### Second example

Showing the two lines not in a single graph but in two separate graphs.

```python
# Initializing the figure
fig2, ax_fig2 = plt.subplots(figsize=(12,5))


# Creating first subplot (third argument) of a grid with one row (first argument) and two columns (second argument)
ax1 = plt.subplot(1,2,1)
plt.plot(x, y1, color="red", marker="o", label="Y1")
plt.title("Test1")
plt.legend()

ax1.set_xticks(x)
ax1.set_xticklabels(month_names)

# Creating second subplot (third argument) of a grid with one row (first argument) and two columns (second argument)
ax2 = plt.subplot(1,2,2)
plt.plot(x, y2 , color ="green", marker ="o", label="Y2")
plt.title("Test2")
plt.legend()

ax2.set_xticks(x)
ax2.set_xticklabels(month_names)



plt.show()

fig2.savefig("My_second_figure.png")
```

#### Third example

Plotting the two variables against each other.

```python
fig3 = plt.figure()

plt.scatter(y1,y2)
#plt.plot(y1, y2,"o")
plt.show()

fig3.savefig("My_third_figure.pdf")
```
